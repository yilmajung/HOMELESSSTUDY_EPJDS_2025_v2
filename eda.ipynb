```python
import geopandas as gpd
import pandas as pd
```

```python
path = "data/main_daily_with_amenities.parquet"
gdf = gpd.read_parquet(path)
print(gdf.shape, gdf.crs)
print(gdf.columns.tolist())
print(gdf.head(3))
```

```python
# To display every column of gdf
pd.option_context("display.max_columns", None)
gdf.head()
```

```python
# histogram
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
gdf['ground_truth'].hist(bins=30)
plt.title('Distribution of Ground Truth Values')
plt.xlabel('Ground Truth')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
gdf['n_atm'].hist(bins=30)
plt.title('Distribution of ATM Counts')
plt.xlabel('Number of ATMs')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
gdf['n_restaurant'].hist(bins=30)
plt.title('Distribution of Restaurant Counts')
plt.xlabel('Number of Restaurants')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
gdf['n_shelter_homeless'].hist(bins=30)
plt.title('Distribution of Homeless Shelter Counts')
plt.xlabel('Number of Homeless Shelters')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```
